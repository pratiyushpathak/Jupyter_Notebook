```python
import numpy as np
import matplotlib.pyplot as plt

ypoints = np.array([3, 8, 1, 10, 5, 7, 2, 9])
xpoints = np.array([2, 7, 0, 9, 4, 6, 1, 8])

plt.plot(ypoints, 'o--y', label = 'Y')
plt.plot(xpoints, 'o--m',  label = 'X')

plt.title('Sports Watch Data')
plt.xlabel('Average Pulse')
plt.ylabel('Calorie Burnage')

plt.show()
```
